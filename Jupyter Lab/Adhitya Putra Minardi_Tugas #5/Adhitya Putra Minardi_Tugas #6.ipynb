```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Jawaban No 1 dan 2

```python
data = pd.read_csv('diabetes_prediction_dataset.csv').dropna()
data
```

## Jawaban No 3 dan 4

```python
data = data.join(pd.get_dummies(data.smoking_history,dtype=float)).drop(['smoking_history'],axis=1)
data = data.join(pd.get_dummies(data.gender,dtype=float)).drop(['gender'],axis=1)

data.drop('No Info',axis=1,inplace=True)
```

## Jawaban No 5

## Jawaban No 6

```python
data.corr().sort_values('diabetes',ascending = False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
LogisticReg = LogisticRegression()
logistic_data = data.loc[:,['diabetes','blood_glucose_level','HbA1c_level','age','bmi']]
x = logistic_data.drop('diabetes',axis = 1)
y = logistic_data['diabetes']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
logistic_data
```

```python
LogisticReg.fit(x_train,y_train)
score = LogisticReg.score(x_test,y_test)
print(score)
LogisticReg.coef_
```

<p>Berdasarkan hasil perhitungan coefficiency diatas dapat disimpulkan bahwa variabel blood_glucose_level memiliki pengaruh paling tinggi dalam menentukan apakah orang tersebut berpotensi diabetes atau tidak </p>

```python
prediction = LogisticReg.predict(x_test)
print(prediction)

data_new = x_test.join(y_test)
data_new['Prediction'] = prediction
data_new.head(10)
# relevant_value = data_new.loc[data['diabetes'] == 1]
# relevant_value
```

```python
from sklearn.metrics import confusion_matrix
new_train = y_train.head(30000)
cm = confusion_matrix(y_test,prediction)
print(cm)
```

```python
import numpy as np
import itertools


def plot_confusion_matrix(cm, classes, title='Confusion Matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')

# Label kelas atau kategori
class_names = [0, 1]  # Misalnya, 0 adalah negatif, dan 1 adalah positif

plt.figure()
plot_confusion_matrix(cm, classes=class_names, title='Confusion Matrix')
plt.show()
```
