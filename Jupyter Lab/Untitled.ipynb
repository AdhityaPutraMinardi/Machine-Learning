```python
import pandas as pd


df = pd.read_csv('breast_cancer.csv')
df
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.3)
y_train = train['Class']
y_test = test['Class']
x_train = train.drop(['Class'], axis=1)
x_test = test.drop(['Class'], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
model = logisticRegr.fit(x_train, y_train)
predictions = logisticRegr.predict(x_test)
print(predictions)
```

```python
x_test['Hasil Prediksi'] = predictions
x_test['Class'] = y_test
x_test
```

```python
score = logisticRegr.score(x_test, y_test)
score
```

```python
x = logisticRegr.predict([[10, 6, 7, 8, 3, 1, 8, 9, 1]])
if x[0] == 4:
    print('Kanker')
else:
    print('Sehat')
```

```python
logisticRegr.coef_
```

```python
from sklearn import metrics
print("Accuracy: ",metrics.accuracy_score(y_test, predictions))
print("Precision : ",metrics.precision_score(y_test, predictions, average='macro'))
print("Recall : ",metrics.recall_score(y_test, predictions, average='macro'))
print("F1-Score : ",metrics.f1_score(y_test, predictions, average='macro'))
print(metrics.classification_report(y_test, predictions))
print(metrics.confusion_matrix(y_test, predictions))
pd.crosstab(y_test, predictions, rownames=['True'], colnames=['Predicted'], margins=True)
```
